## 전기, 가스 데이터 변환

네이버 지오코딩을 돌리기위해 전기, 가스 raw 데이터를 지번 별 사용량 데이터프레임으로 변환

```python
import pandas as pd

path = ""
```

## 가스데이터

```python
gas1 = pd.read_csv(path + 'gas/eais_gas_202201.txt', encoding='cp949', header=None, sep='|')
gas2 = pd.read_csv(path + 'gas/eais_gas_202202.txt', encoding='cp949', header=None, sep='|')
gas3 = pd.read_csv(path + 'gas/eais_gas_202203.txt', encoding='cp949', header=None, sep='|')
gas4 = pd.read_csv(path + 'gas/eais_gas_202204.txt', encoding='cp949', header=None, sep='|')
gas5 = pd.read_csv(path + 'gas/eais_gas_202205.txt', encoding='cp949', header=None, sep='|')
gas6 = pd.read_csv(path + 'gas/eais_gas_202206.txt', encoding='cp949', header=None, sep='|')
gas7 = pd.read_csv(path + 'gas/eais_gas_202207.txt', encoding='cp949', header=None, sep='|')
gas8 = pd.read_csv(path + 'gas/eais_gas_202208.txt', encoding='cp949', header=None, sep='|')
gas9 = pd.read_csv(path + 'gas/eais_gas_202209.txt', encoding='cp949', header=None, sep='|')
gas10 = pd.read_csv(path + 'gas/eais_gas_202210.txt', encoding='cp949', header=None, sep='|')
gas11 = pd.read_csv(path + 'gas/eais_gas_202211.txt', encoding='cp949', header=None, sep='|')
gas12 = pd.read_csv(path + 'gas/eais_gas_202212.txt', encoding='cp949', header=None, sep='|')
```

```python
gas_df = [gas1,gas2,gas3,gas4,gas5,gas6,gas7,gas8,gas9,gas10,gas11,gas12]
```

```python
gas_year = pd.concat(gas_df, axis=0) 
```

```python
gas_year.head(3)
```

```python
new_column_names = [
    "사용년월", "순번", "시군구코드", "법정동코드", "시도명", "시군구명", 
    "법정동명", "대지구분코드", "본번", "부번", "새주소대로로코드", 
    "새주소대로로명", "새주소지상지하코드", "새주소지상지하명", 
    "새주소본번", "새주소부번", "사용량"
]


gas_year.columns = new_column_names

gas_year.head(3)
```

```python
gas_year[['본번', '부번']] = gas_year[['본번', '부번']].astype(int)
gas_year[['본번', '부번']] = gas_year[['본번', '부번']].astype(str)
```

```python
gas_year[['새주소대로로명', '새주소본번', '새주소부번']] = gas_year[['새주소대로로명', '새주소본번', '새주소부번']].astype(str)
```

```python
gas_year["지번"] = gas_year["시도명"] + ' ' +  gas_year["시군구명"] + ' ' + gas_year["법정동명"] + ' ' + gas_year["본번"] + '-' + gas_year["부번"]
```

```python
gas_year["지번"].value_counts()
```

```python
gas_seoul = gas_year[gas_year["지번"].str.contains("서울특별시", na = False)]
```

```python
gas_seoul = gas_seoul.groupby(['지번']).mean()['사용량'].reset_index()
gas_seoul
```

```python
gas_seoul.to_csv(path + "Data/gas_seoul_2022.csv", index=False)
```
